```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv("/kaggle/input/product/statsfinal.csv")
print(data)
```

```python
mean =np.mean(data['Q-P1'])
print(mean)
```

```python
std = np.std(data['S-P1'])
print(std)
```

```python
x = mean
y = std
plt.plot(mean,std)
plt.title("Line graph")
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
total_Q = data['Q-P1'].sum()
total_S = data['S-P1'].sum()
labels = ['Q-P1', 'S-P1']
sizes = [total_Q, total_S]
colors = ['blue', 'red']
explode = (0.1, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Q and S')
plt.show()
```

```python
x = mean
y = std
plt.bar(x,y,color = 'red')
plt.title('Result')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
